```python
import json
from collections import defaultdict
```

```python
with open('responses_prolific.json') as f:
    responses = json.loads(f.read())
    
    total_participants = len(responses)
    passed_attention = 0
    rephrased_preferred = defaultdict(int)
    rephrased_more_natural = defaultdict(int)

    for participant in responses:
        prefer_rephrased = 0
        more_natural_rephrased = 0
        diff_meaning = []
        times = []
        passed = (responses[participant]['11']['subAnswers']['3'] == 'No')
        if passed:
            passed_attention += 1

        for question in responses[participant]:
            if question != '11':
                if responses[participant][question]['subAnswers']['1'] == 'rephrased':
                    prefer_rephrased += 1
                    if passed:
                        rephrased_preferred[question] += 1
                if responses[participant][question]['subAnswers']['2'] == 'rephrased':
                    more_natural_rephrased += 1
                    if passed:
                        rephrased_more_natural[question] += 1
                if responses[participant][question]['subAnswers']['3'] == 'No':
                    diff_meaning.append(question)
            times.append(responses[participant][question]['timeSpent'])
        
        print(participant)
        print('passed attention' if passed else 'failed attention', '- '+responses[participant]['11']['explanation'] if 'explanation' in responses[participant]['11'] else '')
        print()
        
        print('number of rephrased responses preferred:', prefer_rephrased)
        print('number of rephrased responses deemed more natural:', more_natural_rephrased)
        print()

        print('questions with different meanings:')
        for question in diff_meaning:
            if 'explanation' in responses[participant][question]:
                print(f"{question}: {responses[participant][question]['explanation']}")
            else:
                print(f"{question}: no explanation provided")
        print()

        print('min time:  ', min(times))
        print('max time:  ', max(times))
        print('total time:', sum(times))
        print('------------------------')
```

```python
print('rephrased responses preferred:')
for question in rephrased_preferred:
    print(f'{question}: {rephrased_preferred[question]}/{passed_attention} - {rephrased_preferred[question]/passed_attention*100:.2f}%')
print()
print('rephrased responses deemed more natural:')
for question in rephrased_more_natural:
    print(f'{question}: {rephrased_more_natural[question]}/{passed_attention} - {rephrased_more_natural[question]/passed_attention*100:.2f}%')
```
